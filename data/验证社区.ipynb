```python
## 验证社区聚合是否和ixp一一对应

from bs4 import BeautifulSoup

# 指定本地HTML文件路径
file_path = "DE-CIX Frankfurt connected networks.html"  # 替换为您的本地HTML文件路径

# 打开并读取HTML文件
with open(file_path, "r", encoding="utf-8") as file:
    html_content = file.read()

# 使用BeautifulSoup解析HTML
soup = BeautifulSoup(html_content, 'html.parser')

# 找到<tbody>标签
tbody = soup.find('tbody')

# 提取每个<tr>中第一个<td>的<a>标签内容
asn_list = []
if tbody:
    for tr in tbody.find_all('tr'):
        td = tr.find('td')  # 找到第一个<td>
        if td:
            a_tag = td.find('a')  # 找到<td>中的<a>标签
            if a_tag and a_tag.text.isdigit():  # 检查是否为数字
                asn_list.append(int(a_tag.text))  # 提取ASN值并转为整数

# 输出ASN数组
print(asn_list)
```

```python
import pandas as pd
rank = pd.read_csv('202405/handle/rank202405.csv')
rank = rank[rank['id'].isin(asn_list)]
```

```python
rank.to_csv('test.csv')
```
