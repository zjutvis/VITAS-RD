```python
import pandas as pd

rank = None
rel = None
for date in range(202401, 202402):
    # 读取文件
    rank = pd.read_csv(f'./{date}/handle/rank{date}.csv')
    rel = pd.read_csv(f'./{date}/handle/rel{date}.csv')

    # 将 rel 中的 source 和 target 与 rank 合并，生成两列新数据
    rel = rel.merge(rank[['id', 'cone']], left_on='source', right_on='id', how='left').rename(columns={'cone': 'source_cone'}).drop(columns=['id'])
    rel = rel.merge(rank[['id', 'cone']], left_on='target', right_on='id', how='left').rename(columns={'cone': 'target_cone'}).drop(columns=['id'])

    # 比较 source_cone 和 target_cone，保留较大的值
    # rel['cone'] = rel[['source_cone', 'target_cone']].max(axis=1)

    # 如果需要，可以删除中间列
    # rel = rel.drop(columns=['source_cone', 'target_cone'])
```
