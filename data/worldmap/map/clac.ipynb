```python
import geopandas as gpd
import matplotlib.pyplot as plt
import json

# 读取GeoJSON文件
world = gpd.read_file('world_by_region.geojson')

# 调整投影至以中国为中心的亚洲兰伯特等角投影
world_asia_centered = world.to_crs('+proj=laea +lat_0=35 +lon_0=105 +x_0=0 +y_0=0 +ellps=WGS84 +units=m +no_defs')

# 保存为新的GeoJSON文件
world_asia_centered.to_file('world_china_centered.geojson', driver='GeoJSON')

# 创建以中国为中心的世界地图预览
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
world.plot(ax=ax)
ax.set_title('以中国为中心的世界地图')
plt.savefig('world_china_centered_preview.png')
print('已成功创建以中国为中心的GeoJSON文件')
```
