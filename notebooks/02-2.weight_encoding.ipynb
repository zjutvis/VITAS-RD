```python
import pandas as pd
import os
from pathlib import Path

# Periods to process
time_periods = [f"2024{str(i).zfill(2)}" for i in range(1, 11)]

for period in time_periods:
 print(f"\nProcessing period: {period}")
  # Build file paths
 base_path = Path("../visualization/assets/data")
 rank_file = base_path / period / f"rank{period}.csv"
 rel_file = base_path / period / f"rel{period}.csv"

 # Check if files exist
 if not rank_file.exists() or not rel_file.exists():
 print(f"fileexists, skipperiod {period}")
 continue
  try:
 # Read data
 rank_df = pd.read_csv(rank_file, keep_default_na=False)
 rel_df = pd.read_csv(rel_file)
 print(f"Raw data: {len(rank_df)} nodes, {len(rel_df)} relations")
  # IDcolumntype,  rank_df['id'] = rank_df['id'].astype(str)
 rel_df['source'] = rel_df['source'].astype(str)
 rel_df['target'] = rel_df['target'].astype(str)
  # createIDconecountrymapping
 id_to_cone = dict(zip(rank_df['id'], rank_df['cone']))
 id_to_country = dict(zip(rank_df['id'], rank_df['country']))
  # addconefield, sourcetargetcone
 rel_df['source_cone'] = rel_df['source'].map(id_to_cone)
 rel_df['target_cone'] = rel_df['target'].map(id_to_cone)
 rel_df['cone'] = rel_df[['source_cone', 'target_cone']].min(axis=1)

 # addsource_countrytarget_countryfield
 rel_df['source_country'] = rel_df['source'].map(id_to_country)
 rel_df['target_country'] = rel_df['target'].map(id_to_country)
  # deletecolumn
 rel_df = rel_df.drop(['source_cone', 'target_cone'], axis=1)
  # field
 cone_count = rel_df['cone'].notna().sum()
 source_country_count = rel_df['source_country'].notna().sum()
 target_country_count = rel_df['target_country'].notna().sum()
  print(f"process:")
 print(f" - addconefield: {cone_count}/{len(rel_df)} ({cone_count/len(rel_df)*100:.2f}%)")
 print(f" - addsource_countryfield: {source_country_count}/{len(rel_df)} ({source_country_count/len(rel_df)*100:.2f}%)")
 print(f" - addtarget_countryfield: {target_country_count}/{len(rel_df)} ({target_country_count/len(rel_df)*100:.2f}%)")
  # Save processed data
 rel_df.to_csv(rel_file, index=False)
 print(f"Save processed data {rel_file}")
  except Exception as e:
 print(f"Processing period {period} error occurred: {str(e)}")

print("\nAll periods processed")
```

```python
# readtimerange202401-2024-10 reldata, periodprocess
# xxxperiod
# relxxxedgefield, conelog, log(1+cone)/max(log(1+cone)), 0-1
import pandas as pd
import numpy as np
import os
from pathlib import Path

# Periods to process
time_periods = [f"2024{str(i).zfill(2)}" for i in range(1, 11)]

for period in time_periods:
 print(f"\nProcessing period: {period}")

 # Build file paths
 base_path = Path("../visualization/assets/data")
 rel_file = base_path / period / f"rel{period}.csv"

 # Check if files exist
 if not rel_file.exists():
 print(f"fileexists, skipperiod {period}")
 continue

 try:
 # Read data
 rel_df = pd.read_csv(rel_file)
 print(f"Raw data: {len(rel_df)} relations")

 # checkconefieldexists
 if 'cone' not in rel_df.columns:
 print(f"conefieldexists, skipperiod {period}")
 continue

 # log
 log_values = np.log1p(rel_df['cone'])

 # 0-1
 max_log_value = log_values.max()
 rel_df['edge'] = log_values / max_log_value

 #  edge_count = rel_df['edge'].notna().sum()

 print(f"process:")
 print(f" - addedgefield: {edge_count}/{len(rel_df)} ({edge_count/len(rel_df)*100:.2f}%)")
 print(f" - edgerange: {rel_df['edge'].min():.4f} - {rel_df['edge'].max():.4f}")
 print(f" - logrange: {log_values.min():.4f} - {max_log_value:.4f}, : {max_log_value:.4f}")

 # Save processed data
 rel_df.to_csv(rel_file, index=False)
 print(f"Save processed data {rel_file}")

 except Exception as e:
 print(f"Processing period {period} error occurred: {str(e)}")

print("\nAll periods processed")
```

```python
# relfileprocessdirectorytxtformat
# columnsource, columntarget, columnedge()
import pandas as pd
import os
from pathlib import Path

# Periods to process
time_periods = [f"2024{str(i).zfill(2)}" for i in range(1, 11)]

for period in time_periods:
 print(f"\nProcessing period: {period}")
  # Build file paths
 base_path = Path("../visualization/assets/data")
 rel_file = base_path / period / f"rel{period}.csv"
 txt_output_file = base_path / period / "rel1.txt"
  # checkCSVfileexists
 if not rel_file.exists():
 print(f"fileexists, skipperiod {period}")
 continue
  try:
 # Read data
 rel_df = pd.read_csv(rel_file)
 print(f"Raw data: {len(rel_df)} relations")
  # checkcolumnexists
 required_cols = ['source', 'target', 'edge']
 if not all(col in rel_df.columns for col in required_cols):
 print(f"column {required_cols}, skipperiod {period}")
 continue
  # columnformat
 formatted_data = []
 for _, row in rel_df.iterrows():
 # edge
 edge_value = round(row['cone'], 0)
 formatted_data.append(f"{row['source']} {row['target']} {edge_value}")
  # TXTfile
 with open(txt_output_file, 'w', encoding='utf-8') as f:
 f.write('\n'.join(formatted_data))
  print(f"generate {txt_output_file}, {len(formatted_data)} data")
  except Exception as e:
 print(f"Processing period {period} error occurred: {str(e)}")

print("\nAll periods processed")
```
