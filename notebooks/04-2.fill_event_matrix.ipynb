```python
# time1 time2
# xxxtime1
# Read data../visualization/assets/data/xxx/handle/rankxxx.csv
# time1communityfieldid.time2commmunityid
# time1idtime2id, , # : t1_idt2_id/t1_id

def find_most_similar_community(time1, community_number, time2):
 """
 time2time1communitycommunity

 :
 time1: , time ( "2015")
 community_number: , time1community
 time2: , time ( "2016")

 :
 most_similar_comm: time2community
 max_similarity: ()
 """
 # readtime1data
 t1_data = pd.read_csv(f'../visualization/assets/data/{time1}/handle/rank{time1}.csv')

 # readtime2data
 t2_data = pd.read_csv(f'../visualization/assets/data/{time2}/handle/rank{time2}.csv')

 # time1communitynodeID
 t1_community_ids = set(t1_data[t1_data['community'] == community_number]['id'])

 # time1community, None
 if len(t1_community_ids) == 0:
 return None, 0

 # time2community
 t2_communities = t2_data['community'].unique()

 max_similarity = 0
 most_similar_comm = None

 # time2communitytime1community
 for comm in t2_communities:
 t2_community_ids = set(t2_data[t2_data['community'] == comm]['id'])
 intersection = t1_community_ids.intersection(t2_community_ids)

 # : time1community
 similarity = len(intersection) / len(t1_community_ids)

 #  if similarity > max_similarity:
 max_similarity = similarity
 most_similar_comm = comm

 return most_similar_comm, max_similarity
```
