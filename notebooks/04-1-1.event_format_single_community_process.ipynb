```python
import csv
import re


def parse_community_ids(community_str):
 """communityID, communityIDcolumn"""
 if community_str is None or community_str == '':
 return []

 # processformat
 if isinstance(community_str, str):
 community_str = community_str.strip()
 if not community_str:
 return []

 # processcolumnformat "[8, 35]"
 if community_str.startswith('[') and community_str.endswith(']'):
 return [id.strip() for id in re.findall(r'\d+', community_str)]
 else:
 return [community_str]

 # processtype
 if isinstance(community_str, (int, float)):
 return [str(int(community_str))]

 return []


def track_community_evolution(start_date, start_community_id, csv_file_path):
 """
 community, 
 :
 start_date (str): time, formatYYYYMM
 start_community_id: communityID
 csv_file_path (str): CSVfilepath

 :
 str: communitypath
 """
 # readCSVdata
 events = []
 with open(csv_file_path, 'r', encoding='utf-8') as file:
 reader = csv.DictReader(file)
 for row in reader:
 events.append(row)

 # event
 events_by_date = {}
 all_dates = set()
 for event in events:
 source_date = event['source_date']
 if source_date not in events_by_date:
 events_by_date[source_date] = []
 events_by_date[source_date].append(event)
 all_dates.add(source_date)
 all_dates.add(event['target_date'])

 # time
 sorted_dates = sorted(list(all_dates))

 #  current_date = str(start_date)
 current_communities = {str(start_community_id): None} # communityID -> communityID
 evolution_path = [str(start_community_id)]

 #  date_index = sorted_dates.index(current_date) if current_date in sorted_dates else -1

 while date_index >= 0 and date_index < len(sorted_dates) - 1:
 current_date = sorted_dates[date_index]
 next_date = sorted_dates[date_index + 1]
 next_communities = {}

 # processevent
 if current_date in events_by_date:
 for event in events_by_date[current_date]:
 if event['target_date'] != next_date:
 continue

 source_ids = parse_community_ids(event['source_community'])

 # checkcommunity
 for source_id in source_ids:
 if source_id in current_communities:
 # targetcommunity
 target_ids = parse_community_ids(event['target_community'])

 # process, ,  if event['event_type'] == '':
 for target_id in target_ids:
 next_communities[target_id] = source_id
 elif event['event_type'] == '' and len(source_ids) > 1:
 # processcommunity
 if source_id in current_communities:
 for target_id in target_ids:
 next_communities[target_id] = source_id
 elif event['event_type'] == '':
 # community, addnext_communities
 pass
 else: # , add,  for target_id in target_ids:
 next_communities[target_id] = source_id

 # path
 if next_communities:
 # community, targetcommunity
 source_to_targets = {}
 for target, source in next_communities.items():
 if source not in source_to_targets:
 source_to_targets[source] = []
 source_to_targets[source].append(target)

 #  next_step = []
 for source_id in current_communities:
 if source_id in source_to_targets:
 targets = source_to_targets[source_id]
 if targets:
 if len(targets) > 1:
 next_step.append(f"[{','.join(targets)}]")
 else:
 next_step.append(targets[0])
 else:
 next_step.append("") # community
 else:
 next_step.append("") # community

 # addpath
 if len(next_step) == 1:
 evolution_path.append(next_step[0])
 else:
 # [,]format
 combined = []
 has_empty = any(s == "" for s in next_step)
 has_value = any(s != "" for s in next_step)

 if has_empty and has_value:
 # [,][,]format
 values = [s for s in next_step if s]
 if len(values) == 1:
 evolution_path.append(f"[,{values[0]}]")
 else:
 evolution_path.append(f"[{','.join(values)}]")
 elif has_value:
 evolution_path.append(f"[{','.join(next_step)}]")
 else:
 evolution_path.append("") #  else:
 evolution_path.append("") # 
 #  current_communities = next_communities
 date_index += 1

 # community,  if not current_communities:
 break

 # path
 return ','.join(evolution_path)


# 1: 202401community3
result1 = track_community_evolution('202401', 3, "../visualization/assets/data/events/community_events.csv")
print(f"community3202401path: {result1}")

# 2: 202401community1
result2 = track_community_evolution('202401', 15, "../visualization/assets/data/events/community_events.csv")
print(f"community1202401path: {result2}")
```
